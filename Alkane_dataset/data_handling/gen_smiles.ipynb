This jupiter notebook is for generating the smile srings for all the compounds except the n-alkanes.
To achieve this the python package CIRpy was used.

```python
import pandas as pd
import cirpy
```

```python
df = pd.read_csv('../csv_files/PE_2_3_combined_raw_new.csv')

#keep trace on which identifier was the most efficient
name_found = 0
casnr_found = 0
exceptions = 0

smiles = []

#we have 802 samples is this df
for i in range(df.shape[0]):
    if i in [100, 200, 300, 400, 500, 600, 700]:
        print("genereting smiles string for compound", i)

    comp_name = df['Component name'][i]
    casnr = df['CAS number'][i]

    #exceptionnally cirpy returns an error than jut add none
    try: 
        smile = cirpy.resolve(comp_name, 'smiles')
    except:
        smile = None
    try: 
        smile2 = cirpy.resolve(casnr, 'smiles')
    except:
        smile2 = None
    if smile == None and smile2 == None:
        exceptions +=1 

    #use some if statements to add the rigth value
    #if both retun None or the rigth smiles string add it
    if smile == smile2:
        smiles.append(smile)
        name_found +=1
        casnr_found +=1
    
    # if they both return a diffrent smile add the smile string generated from the casnr
    elif smile != None and smile2 != None:
        smiles.append(smile2)
    
    #if only smile is None add smile2
    elif smile == None:
        smiles.append(smile2)
        casnr_found +=1

    else:
        smiles.append(smile)
        name_found +=1

```

```python
print("amount found by compound name:", name_found)
print("amount found by casnr:", casnr_found)
print("amount of errors", exceptions)
```

```python
df= pd.read_csv('../csv_files/PE_2_3_combined_raw_new.csv')
```

```python
df.insert(3,"smiles",smiles,True)
#df.to_csv('../csv_files/PE_2_3_combined_final.csv')
```

```python
data = df.drop(columns=["Unnamed: 0","CAS number", "Component name"])
idx = []

#find the molecules which did not return a smile string
for i, smile in enumerate(data.smiles):
    if smile == None:
        idx.append(i)

#remove them
data = data.drop(idx)
```
