```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../csv_files/CombinedDataset.csv")
```

```python
molwt = df["Mw"].tolist()
```

```python
plt.hist(molwt)
plt.xlabel("Molecular weight (g/mol)")
plt.ylabel("Amount of molecules (a.u.)")
plt.title("Molecular weight distribution of the alkanes dataset")
plt.show()
```
