```python
import pandas as pd
```

```python
df = pd.read_csv("../csv_files/Combined_dens_Dataset.csv")
df['state']=0
df
```

```python
for i in range(df.shape[0]):
    dens = df.iloc[i]['density']

    # if it is a gas
    if dens <= 13:
        state = 0
    
    # if it is between gas and liquid
    elif 13 < dens < 620:
        state = 1
    
    # if it is a liquid
    else:
        state = 2
    
    df.at[i,"state"] = state

df
```

```python
df.to_csv("../csv_files/Combined_dens_Dataset_WithState.csv")
```
