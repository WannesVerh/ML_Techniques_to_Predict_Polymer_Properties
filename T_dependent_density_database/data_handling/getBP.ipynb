```python
import pandas as pd
```

```python
df = pd.read_csv("../csv_files/AspenProp_Combined_dataset.csv")
df
```

```python
df2 = pd.read_csv("../csv_files/Combined_dens_Dataset.csv")
df2
```

```python
df2['TB']= 0
df2
```

```python
for i in range(df2.shape[0]):
    smile = df2.iloc[i]['smiles']
    try:
        index = df[df['smiles'] == smile].index[0]
        Tb = df.iloc[index]['TB']
    except:
        Tb = 0
    
    
    df2.at[i,'TB'] = Tb
```

```python
df2 = df2[df2.TB != 0]
```

```python
df2.to_csv("../csv_files/Combined_dens_Dataset_WithTB.csv", index=False)
```
